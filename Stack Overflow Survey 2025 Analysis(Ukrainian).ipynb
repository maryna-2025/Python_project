### Завдання 1. Підрахунок загальної кількості респондентів

```python
#  .nunique() - метод series, який підрахову кількість унікальних значень у стовпці ResponseID
unique_respondents = df['ResponseId'].nunique()
print( f'Загальна кількість респондентів: {unique_respondents}.')
```

### Завдання 2. Аналіз повноти відповідей респондентів

```python
# questions - створюємо множину назв питань, які присутні одночасно в qname та серед ствопців df.
# використання set() та qname.unique() створює множину унікальних qname,
names_in_df1 = set(df1['qname'].unique())
#  після чого ми беремо перетин із df.columns, щоб отримати лише ті питання, які реально існують в датафреймі.
all_columns = set(df.columns)
questions = names_in_df1 & all_columns
# dropna(): видалає тільки ті рядки де є промуски (NaN) у зазначиних стовпцях.
df_questions = df[list(questions)]
df_questions_nonull = df_questions.dropna()
# shape[0] повертає кільуість рядків - тобто кількість респондентів, які відповіли на всі ці питання.
q = df_questions_nonull.shape[0]
print(f'Кількість респондентів, які відповіли на всі ці питання: {q}.')
```

```python
#перевіряємо результат
rows_with_nulls = df_questions.isnull().any(axis=1).sum()
print(f"Кількість респондентів всього: {df_questions.shape[0]}")
print(f"Кількість респондентів з пропусками: {rows_with_nulls}")
print(f'Кількість респондентів, які відповіли на всі ці питання: {q}.')
```

### Завдання 4. Аналіз віддаленої роботи 

```python
#  Підраховуємо кількість рядків у df, де значення у відповідній колонці дорівнює "Remote". 
remote_workers = df['RemoteWork'].str.contains('Remote', case=False, na=False).sum() 
print(f'Кількість респондентів, які працюють віддалено: {remote_workers}.')
```

### Завдання 5. Визначення популярності Python

```python
# Створюємо новий стовпець worked_with_python, який містить Ture/False,
# де True означає, що респондент згадував Python у LanguageHaveWorkedWith.
# str.contains('...') - шукаємо слово 'python'
#  без урахування регісту та змінюємо пропуски (NaN) на False.
df['worked_with_python'] = df['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False)
```

```python
# sum() - рахуємо загальну кількість респондентів, які працювали з Python.
# - обчислюємо частку таких респондентів у відсоткпх
worked_with_python = df['worked_with_python'].sum()
all_respondents = df['LanguageHaveWorkedWith'].count()
worked_with_python_percents = (worked_with_python / all_respondents) * 100
print(f'Відсоток респондентів які працювали з Python: {worked_with_python_percents: .2f}%')
```

### Завдання 7. Географічний аналіз компенсації Python-розробників

```python
# df.worked_with_python — відбираємо лише тих, хто працював з Python.
# Видаляємо рядки, де відсутнє значення зарплати (ConvertedCompYearly).
# .groupby() — групуємо дані за країною (Country) і обчислюємо середню (mean) та медіанну (median) зарплату для кожної країни.
python_dev_compen = df.loc[(df['worked_with_python'] == True) & (df['ConvertedCompYearly'].notnull()),['Country', 'ConvertedCompYearly']]
py_progers_by_country = python_dev_compen.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median'])
py_progers_by_country_sort = py_progers_by_country.sort_values(by='median', ascending=False)
```

```python
# py_progers_by_country_sort — таблиця з результатами.
py_progers_by_country_sort
```

### Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів

```python
# Вибираємо три стовпці: ResponseId, EdLevel, ConvertedCompYearly.
# Сортуємо за колонкою ConvertedCompYearly у порядку спадання (від більшої зарплати до меншої).
short_df = df[['ResponseId', 'EdLevel', 'ConvertedCompYearly']].sort_values(by='ConvertedCompYearly', ascending=False)

# reset_index(drop=True) — оновлюємо індекси після сортування, старий індекс видаляємо.
best_comp_df = short_df.reset_index(drop=True)

# head(5) — відбираємо перші 5 рядків (топ-5 респондентів з найвищою зарплатою).
best_comp_df.head(5)
```

### Завдання 9 Аналіз популярності Python по віковим категоріям 

```python
# Рахуємо середнє значення колонки 'worked_with_python' для кожної вікової групи
# Це дає частку, яку ми множимо на 100 для отримання відсотків
python_by_age = df.groupby('Age')['worked_with_python'].mean() * 100

print('Популярність Python по віковим категоріям:')
print(python_by_age.round(2))
```

### Завдання 10 Аналіз індустрій серед високооплачуваних віддалених працівників

```python
# Знаходимо респондентів, які знаходяться в 75 перцентилі 
df_comp75 = df['ConvertedCompYearly'].quantile(0.75)
df_high_comp75 = df[df['ConvertedCompYearly'] >= df_comp75]

# відбираємо тільки тих, хто працює віддалено
mask = df_high_comp75['RemoteWork'].str.contains('Remote', case=False, na=False)
p75_on_remote = df_high_comp75[mask]

# визначаємо найрозповсюдженніші індустрії
industru_count = p75_on_remote.groupby('Industry')['Industry'].count()
top_indust_on_remote = industru_count.sort_values(ascending=False)
top_indust_on_remote
```
